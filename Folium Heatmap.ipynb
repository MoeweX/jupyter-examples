```python
# Imports

import folium
from folium.plugins import HeatMap
import os
import pandas
```

```python
# Read in indices files

dataDir = os.path.abspath("./Data/Map Data/")
print("Reading from " + dataDir)

indices = pandas.read_csv(dataDir + '/indices.csv', delimiter=";", names=["index", "path"])
indices.head()
```

```python
# Read in data and plot on map

# Overview

MAX_INDEX = 20
ENTRIES = 9999999
DIVISOR = 1

heat_data = []

for i in range(1, MAX_INDEX + 1):
    filePath = dataDir + "/" + indices.iloc[i - 1]["path"]
    
    with open(filePath, "r") as f:
        for i, line in enumerate(f):
            if (i > 6 and i < 6 + ENTRIES and i % DIVISOR == 0) or i == 6:
                split = line.split(",")
                lat = float(split[0])
                lon = float(split[1])
                heat_data.append([lat, lon])
                                
m = folium.Map(location=[39.9, 116.4], zoom_start=11, tiles="Mapbox Bright", control_scale = True)
print("There are {0!s} entries on the map".format(len(heat_data)))
hm = HeatMap(heat_data, radius=10)
hm.add_to(m)
# m.save("overview-map.html")
m
```
